```python
act = pd.read_csv("../data/activity_cleared_with_corpus.csv",
                  index_col=0).reset_index()
del act['index']
act
```

```python
temp = act[(act['city_name'] == '베네치아') | (act['city_name'] == '산티아고')]
temp
```

```python
# 상품 제목, 연령 추가 크롤링
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
```

```python
text = []

for act in temp['activity']:
    driver = webdriver.Chrome()
    driver.get("https://www.myrealtrip.com/offers/{}".format(act))
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    title = driver.find_element_by_css_selector('#OfferPage__top').text
    try:
        kind = driver.find_element_by_css_selector(
            '#review > div.offer-review__detail-info > div.offer-review__detail-info__stats > p').text
    except:
        kind = '없음'
        
    city_name = driver.find_element_by_css_selector(
        '#offer > div.offer-container__inner-bar--wrapper > div.offer-container__inner-bar--sub-wrapper > p.offer-container__inner-bar--city > a').text
    text.append({'city_name': city_name, 'activity': act,
                 'tour_name': title, 'kind': kind})

    driver.quit()
```
